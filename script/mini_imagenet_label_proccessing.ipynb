```python
import os
import pandas as pd
import json

root= '/home/royliu/Documents/datasets'

mini_label_path = os.path.join(root,'./mini_imagenet/labels.json')
mini_label_dict = json.load(open(mini_label_path, "r"))


label_path =os.path.join(root,'./imagenet1000_labels.txt')
label_dict = pd.read_table(label_path, header=None, index_col = None)
label_dict

#generate a dataframe of imagenet 1000 labels
imagenet_label = pd.DataFrame(columns=['class_value','class_name'])
mini_imagenet_label = pd.DataFrame(columns=['class_value','class_name','label'])
for i, label_class in enumerate(label_dict[0]):
    class_value,class_name = label_class.split(":")
    class_name = class_name[2:-2]  #.strip(",").strip("\'")
    imagenet_label.loc[i]= [int(class_value),class_name]

# generate 
for  i, item in enumerate(mini_label_dict.items()):
    label, class_name = item
    record = imagenet_label.where(imagenet_label['class_name']== class_name).dropna()
    class_value = int(record['class_value'].to_numpy()[0])

    mini_imagenet_label.loc[i]=[class_value,class_name, label]
mini_imagenet_label.to_csv(os.path.join(root, './mini_imagenet/mini_imagenet_labels.csv'),index= False)

```
